```python
import pandas as pd
from openai import OpenAI
import threading
```

```python
def update_cell(df, col, row_idx,client):
    # i = 0
    new_text = ""
    original_text = df.at[row_idx, col]
    prompt = f"I want to convert the data into raw string format for my React app. Do not modify any word in the text. Apply basic formatting such as adding newline and double asterisks for making important titles bold. If there are no titles, do not apply bold formatting. Provide the raw text. {original_text}"
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt }],
        stream=True,)
    for chunk in response:
        if chunk.choices[0].delta.content is not None:
            # print(f'{i} time')
            new_text += f'{chunk.choices[0].delta.content}'
            # print(new_text)
            # i = i+1
            
    # new_text = response.choices[0].text.strip()
    print(f"{row_idx , new_text} replaced")
    df.at[row_idx, col] = new_text
    
```

```python
def update_column(df, col,client):
    for row_idx in df.index:
        print(row_idx)
        update_cell(df, col, row_idx,client)
```

```python
def main(file_path, columns, output_file_path,client):
    df = pd.read_excel(file_path)
    for col in columns:
        update_column(df,col,client)
    print("Done")
    df.to_excel(output_file_path, index=False)
```

```python
if __name__ == "__main__":
    api_key = '' 
    client = OpenAI(api_key=api_key)
    file_path = 'lta.xlsx' 
    columns = ['Lesson Plan']
    output_file_path = 'output.xlsx' 
    main(file_path, columns, output_file_path,client)
```
