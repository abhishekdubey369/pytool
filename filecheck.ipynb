```python
import pandas as pd
from threading import Thread,Lock
```

```python
lock = Lock()
```

```python
data = pd.read_excel("ltass.xlsx")
```

```python
df = data.dropna()
```

```python
df.columns
```

```python

def update_cell(df, col, row_idx):
    original_text = df.at[row_idx, col]
    # Ensure thread-safe update
    with lock:
        df.at[row_idx, col] = original_text
        print(f"Row {row_idx}, column '{col}' updated",end=" ")
```

```python
def update_column(df,row,col):
    for row_idx in row:
        update_cell(df, col, row_idx)
        df[col] = df[col].apply(lambda x: x.replace('\n', '\\n') if isinstance(x, str) else x)
```

```python
columns = [col for col in df.columns]
    # row = [0,1,5,6,8,9,12,13,15,18,19,20,21,22,23,26,27,28,29,30,33,34,35,36,39,40,41,42,43,46,50,51,53,54,55,56,57,58,60,61,62,64,65,66,67,68,69]
row = [r for r in df.index]
threads = []  

for col in columns:
    thread = Thread(target=update_column, args=(df,row, col))
    threads.append(thread)
    thread.start()
    
for thread in threads:
    thread.join()

# Use xlsxwriter to write the DataFrame to Excel
writer = pd.ExcelWriter("hello.xlsx", engine='xlsxwriter')
df.to_excel(writer, index=False, sheet_name='Sheet1')
writer.close()
print("\n\nProcessing complete and output file saved")
```
